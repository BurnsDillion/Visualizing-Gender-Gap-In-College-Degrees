# Analyzing Gender Gap In College Degrees

This data is from the Department of Education. They release a data set annually containing percentage of Bachelor's degree granted. The dataset we will be using is cleaned by a data scientist by the name of Randal Olson. 

```python
# Opening the Data
# %matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

# Data Exploration

```python
print(women_degrees.info())
```

Above we can see the columns are categories of College Degrees (17 of them) and 1 column as the year, displaying 42 years.
For now, we will be choosing particular STEM careers listed below:
- Engineering
- Computer Science
- Psychology
- Biology
- Physical Sciences
- Math and Statistics

```python
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
cb_dark_blue = (0/255, 107/255, 164/255)
cb_orange = (255/255, 128/255, 14/255)
```

Creating 3 Loops to plot the three categories of degrees.

```python
fig = plt.figure(figsize=(17, 20))

for sp in range(0,18,3):
    cat_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(stem_cats[cat_index])
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.5)
    
    
    if cat_index == 0:
        ax.text(2007, 87, 'Men')
        ax.text(2005, 8, 'Women')
    elif cat_index == 5:
        ax.text(2007, 62, 'Men')
        ax.text(2005, 35, 'Women')
        ax.tick_params(labelbottom=True)
        
    
for sp in range(1,16,3):
    cat_index = int((sp-1)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(lib_arts_cats[cat_index])
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.5)
    
    if cat_index == 4:
        ax.tick_params(labelbottom=True)
        
    if cat_index == 0:
        ax.text(2005, 74, 'Women')
        ax.text(2007, 22, 'Men')
        
for sp in range(2,19,3):
    cat_index = int((sp-2)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(other_cats[cat_index])
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.5)
       
    
    if cat_index == 0:
        ax.text(2005, 90, 'Women')
        ax.text(2007, 19, 'Men')
    elif cat_index == 5:
        ax.text(2007, 62, 'Men')
        ax.text(2005, 35, 'Women')
        ax.tick_params(labelbottom=True)
        
        
plt.savefig('gender_degrees.png')        
plt.show()
```

Above you can see all of the categories of college degrees in the data along with the gender percentage gap within them over time. The horizontal line notates the `50%` mark.
